Tushare is a free data api. You can get comprehensive financial data of Chinese market from it. The official website of Tushare is

After you get your token, you can start requesting data from Tushare. Generally, you can only request for daily stock market OHLC data at initial. If you want to access to more kinds of data, you need to contribute to the community.

Now let's start!

```python
import tushare as ts
import pandas as pd
```

```python
# ts.set_token('your token here') 
```

```python
# Initialize the data interface
ts_pro_api = ts.pro_api()
```

<font color=red> If you are not familiar with the Chinese stock market, you can get basic information of each stock by "stock_basics"

```python
# Acquire basic information of stocks. ts_code is the unique code of the stock and is used to request more detailed information of this stock.
data = ts_pro_api.stock_basic(exchange='', list_status='L', fields='ts_code,symbol,name,area,industry,list_date')
data.head(10)
```

<font color=red> Then try to get the daily open,high,low,close data of 000001.SZ from 2015-01-01 to 2022-01-01

```python
df_000001_SZ = ts_pro_api.daily(ts_code='000001.SZ', start_date='20150101', end_date='20220101')
df_000001_SZ.tail(10)
```

<font color=red> In the following research, we are likely to process the data of the whole market. But it is annoying to request data of each stock one by one. Once we have the basic information of all the stocks, we can tranlate the ts_code series to a string containing all the stock codes, and request the data of all stocks in a pack. But be aware that we can only request maximum of 1000 stocks data in a pack. The next is an example of how to get the daily trading data of 1000 stocks from 2021-01-01 to 2022-01-01.

```python
# Translate the ts_code series into a string
# This method can request maximum of 1000 stocks data in a pack
code_list = data['ts_code'].tolist()
code_str = ''
for i in range(0,999,1):
    code = code_list[i] + ','
    code_str += code
code_str += code_list[999]
code_str[0:100]
```

```python
df = ts_pro_api.daily(ts_code= code_str, start_date='20210101', end_date='20220101')
df.head(30)
```

<font color=red> We can also request the one day snapshot of the whole market. See the example below

```python
snapshot = ts_pro_api.daily(trade_date='20220401')
snapshot.tail(30)
```
